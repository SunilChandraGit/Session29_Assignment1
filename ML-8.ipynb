```python
import nltk
from nltk import word_tokenize
#nltk.download()
```

```python
import urllib.request
from bs4 import BeautifulSoup

#read the webpage
response = urllib.request.urlopen('http://php.net/')
html = response.read()
soup = BeautifulSoup(html,"html5lib")
```

```python
#get the stopwords
#nltk.download('stopwords')
sw = nltk.corpus.stopwords.words('english')
```

```python
#get html content as text
text = soup.get_text()
```

```python
#filter out words
import re
tokens = re.findall('\w+', text)
```

```python
#convert all words to lowercase
words = []

for word in tokens:
    words.append(word.lower())
```

```python
#remove all the stopwords
word_ns = []

for word in words:
    if word not in sw:
        word_ns.append(word)
```

```python
#remove all the digits
nlp_dist_no_integers = [x for x in word_ns if not (x.isdigit() or x[0] == '-' and x[1:].isdigit())]
nlp_dist = nltk.FreqDist(nlp_dist_no_integers)
```

```python
#plot frquency of words
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set({'figure.figsize':(23.4,8.27)})
sns.set_style('darkgrid')

nlp_dist.plot(50)
```
